# Dropbox VS competidores

En éste ejerciciop voy a comparar los diferentes planes económicos de las plataformas que ofrecen almacenamiento en la nube, con el fin de compararlos con el plan de Dropbox.

Voy a interpretar los servicios que ofrecen exclusivamente el servicio de almacenamiento en la nube como competidores directos, y los que ofrecen muchos más servicios, ya sea de forma directa o integrado dentro de otro sistema, (como las notas de Notion), como competidores indirectos.

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Mapas de color y marcador
color_map = {"Dropbox": "blue", "Directo": "red", "Indirecto": "orange"}
marker_map = {"Dropbox": "D", "Directo": "o", "Indirecto": "s"}

# Planes de pago de cada servicio
planes = [
    ("Dropbox", 9.99, 2000, "Dropbox"),         # Dropbox Plus
    ("Google Drive", 13.60, 2000, "Directo"),   # Business Starter
    ("OneDrive", 99/12, 1000, "Directo"),       # Microsoft 365 Básico
    ("iCloud", 9.99, 2000, "Directo"),          # Apple iCloud
    ("pCloud", 9.99, 2000, "Directo"),          # Premium 500GB
    ("Sync.com", 6, 1000, "Directo"),           # Solo Professional
    ("Mega", 9.99, 2000, "Directo"),            # Pro I
    ("Tresorit", 9.99, 1000, "Directo"),        # Personal Lite
    ("Zoho WorkDrive", 2.50, 1000, "Directo"),  # Starter

    ("Canva", 110/12, 1000, "Indirecto"),       # Canva Pro
    ("Adobe", 14.74, 1000, "Indirecto"),        # Adobe Lightroom
    ("Frame.io", 15, 2000, "Indirecto"),        # Frame Pro
    ("Notion", 9.50, 2000, "Indirecto"),        # Plus
    ("Avid", 99/12, 500, "Indirecto"),          # Avid Producer Cloud Plan
    ("LabArchives", 330/12, 100, "Indirecto")   # LabArchives ELN Professional
]

# Separar datos
nombre = [x[0] for x in planes]
precio_m = [x[1] for x in planes]
gb = [x[2] for x in planes]

# Gráfico
plt.figure(figsize=(10, 8))
superposicion = defaultdict(int)

for nombre, precio, gb, tipo in planes:
    plt.scatter(gb, precio, color=color_map[tipo], marker=marker_map[tipo], s=100)

    # Contador de cuántas etiquetas hay ya en este punto (gb, precio)
    count = superposicion[(gb, precio)]
    offset_y = 0.7 * count  # Separación etiquetas vertical
    plt.text(gb + 50, precio + offset_y, nombre, fontsize=15, color=color_map[tipo])
    superposicion[(gb, precio)] += 1  # Actualizar contador

# Configuraciones
plt.title("Mapa de dispersión Dropbox VS competidores")
plt.xlabel("Almacenamiento (GB)")
plt.ylabel("Precio mensual (EUR)")
plt.grid(True)
plt.xlim(0, 2250)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
superposicion = defaultdict(int)

for nombre, precio, gb, tipo in planes:
    ratio = gb / precio  # GB por euro
    count = superposicion[(ratio, precio)]
    offset_y = 0.7 * count  # Separación etiquetas vertical

    plt.scatter(ratio, precio, color=color_map[tipo], marker=marker_map[tipo], s=100)
    plt.text(ratio + 5, precio + offset_y, nombre, fontsize=15, color=color_map[tipo])
    superposicion[(ratio, precio)] += 1

plt.title("Relación cantidad-precio (GB por euro)")
plt.xlabel("Ratio almacenamiento / precio (GB por EUR)")
plt.ylabel("Precio mensual (EUR)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Conclusion
Viendo la relación de precio por GB, podemos ver que muchos de los "competidores directos" de DropBox no ofrecen mejores precios por BG que DropBox, lo que prueba que su plan sí es potencialmente atractivo para muchos usuarios.

También podemos apreciar que los competidores "indirectos" ofrecen en algunas ocasiones una buena relación cantidad precio, por lo que puede ser interesante optar por estos servicios si sus funciones se adaptan a nuestras necesidades. Pero también hay que destacar que para los servicios de Adobe, Avid, y sobretodo LabArchives, no es el caso, ya que tienen un elevado precio y están muy orientadas a un uso muy concreto, a no ser que ya usemos otros de sus productos y nos resulte conveniente su integración dentro de un mismo ecosistema.
